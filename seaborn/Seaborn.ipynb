```python
import seaborn as sns
import pandas as pd
import warnings
```

```python
dataset = pd.read_csv("PrePlacement.csv")
dataset
```

```python
warnings.filterwarnings("ignore")
dataset.shape
```

```python
#Taking only Placed Students data
dataset  = dataset[dataset["status"] == "Placed"]
dataset
```

```python
sns.distplot(dataset["ssc_p"],hist=True)
```

```python
sns.jointplot(x='ssc_p',y = 'salary', data = dataset)
```

```python
sns.jointplot(x='ssc_p',y = 'salary', data = dataset,kind = "kde")
```

```python
sns.jointplot(x='ssc_p',y = 'salary', data = dataset,kind = "hex")
```

```python
#Pair Plot
sns.pairplot(dataset, hue = "gender" , diag_kind = "kde", kind = "scatter", palette = "husl" )
```

# For Categorical Plot

```python
sns.stripplot(x = "ssc_b", y = "ssc_p", data = dataset)
```

```python
sns.stripplot(x = "gender", y = "ssc_p", data = dataset)
```

```python
sns.swarmplot(x = "ssc_b", y = "ssc_p", data = dataset)
```

# Distribution Observation

```python
sns.violinplot(x = "gender", y = "ssc_p", data = dataset)
```

```python
sns.violinplot(x = "gender", y = "ssc_p", data = dataset, hue = "gender")
```

```python
sns.violinplot(x = "specialisation", y = "salary", data = dataset, hue = "gender")
```

```python
sns.boxplot(x = "gender", y = "ssc_p", data = dataset, hue = "gender")
```

```python
sns.barplot(x="ssc_b",y = "ssc_p",data = dataset,hue = "gender")
```

```python
sns.countplot(x="ssc_b",data = dataset,hue = "gender")
```

```python
sns.pointplot(x="ssc_p", y = "hsc_b",data = dataset,hue = "gender")
```
